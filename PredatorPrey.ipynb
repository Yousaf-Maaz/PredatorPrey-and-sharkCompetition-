```python
#This code performs the necessary steps to prepare to plot.
from cmath import sqrt
from pylab import *     #Import plotting module matplotlib as well as other modules
                        #into the global namespace.

#Set the default plot to be inline rather than a new window.
%matplotlib inline

import inspect  #This allows us to print the source code in this notebook.
from PredatorPrey import *  #Import the custom functions from this project.

goldenratio=1/2*(1+sqrt(5))        #The next few lines are used for the size of plots
fsx=7                              #Width (in inches) for the figures.
fsy=fsx/goldenratio                #Height (in inches) for the figures.
import time
from PredatorPrey import PredatorPrey
```

```python
#Simulate
timeLst, predatorLst, preyLst = PredatorPrey()
```

```python
# Call RK-2 Implementation for Predator-Prey Model
start_time = time.time()
tLst_RK2, predatorLst_RK2, preyLst_RK2 = PredatorPrey(DT=0.001, simLength=12, method='RK2')
end_time = time.time()
execution_time_RK2_PP = end_time - start_time

# Call RK-4 Implementation for Predator-Prey Model
start_time = time.time()
tLst_RK4, predatorLst_RK4, preyLst_RK4 = PredatorPrey(DT=0.001, simLength=12, method='RK4')
end_time = time.time()
execution_time_RK4_PP = end_time - start_time

# Assuming Euler is the default method, we'll measure its execution time as well
start_time = time.time()
tLst_Euler, predatorLst_Euler, preyLst_Euler = PredatorPrey(DT=0.001, simLength=12)
end_time = time.time()
execution_time_Euler_PP = end_time - start_time
```

```python
from pylab import *

# Plotting the population of predators and prey over time
figure(figsize=(7, 5))
plot(tLst, predatorLst, 'b-', label='Predator Population')
plot(tLst, preyLst, 'r-', label='Prey Population')
legend(loc='upper center')
xlabel('Time')
ylabel('Population')
title('Predator and Prey Populations vs Time')

show()
```

```python
# Print execution times for Predator-Prey Model
print(f"Execution time for RK-2 Predator-Prey: {execution_time_RK2_PP} seconds")
print(f"Execution time for RK-4 Predator-Prey: {execution_time_RK4_PP} seconds")
print(f"Execution time for Euler Predator-Prey: {execution_time_Euler_PP} seconds")
```
