### Species competition between white tip sharks (WTS) and black tip sharks (BTS)

```python
#This code performs the necessary steps to prepare to plot.
from cmath import sqrt
from pylab import *     #Import plotting module matplotlib as well as other modules
                        #into the global namespace.

#Set the default plot to be inline rather than a new window.
%matplotlib inline

import inspect  #This allows us to print the source code in this notebook.
from sharkCompetition import *  #Import the custom functions from this project.

goldenratio=1/2*(1+sqrt(5))        #The next few lines are used for the size of plots
fsx=7                              #Width (in inches) for the figures.
fsy=fsx/goldenratio                #Height (in inches) for the figures.
import time
from sharkCompetition import sharkCompetition
```

```python
#Simulate
timeLst, WTSLst, BTSLst = sharkCompetition()

# Call RK-2 Implementation for Shark Competition Model
start_time = time.time()
tLst_RK2_shark, WTSLst_RK2, BTSLst_RK2 = sharkCompetition(DT=0.001, simLength=5, method='RK2')
end_time = time.time()
execution_time_RK2_shark = end_time - start_time

# Call RK-4 Implementation for Shark Competition Model
start_time = time.time()
tLst_RK4_shark, WTSLst_RK4, BTSLst_RK4 = sharkCompetition(DT=0.001, simLength=5, method='RK4')
end_time = time.time()
execution_time_RK4_shark = end_time - start_time

# Assuming Euler is the default method, we'll measure its execution time as well
start_time = time.time()
tLst_Euler_shark, WTSLst_Euler, BTSLst_Euler = sharkCompetition(DT=0.001, simLength=5)
end_time = time.time()
execution_time_Euler_shark = end_time - start_time
```

```python
from turtle import title
from pylab import *

# Plotting the population of WTS and BTS over time
figure(figsize=(7, 5))
plot(tLst_shark, WTSLst, 'b-', label='WTS Population')
plot(tLst_shark, BTSLst, 'r-', label='BTS Population')
legend(loc='upper center')
xlabel('Time')
ylabel('Population')
title('WTS and BTS Populations vs Time')

show()
```

```python
# Print execution times for Shark Competition Model
print(f"Execution time for RK-2 Shark Competition: {execution_time_RK2_shark} seconds")
print(f"Execution time for RK-4 Shark Competition: {execution_time_RK4_shark} seconds")
print(f"Execution time for Euler Shark Competition: {execution_time_Euler_shark} seconds")
```
